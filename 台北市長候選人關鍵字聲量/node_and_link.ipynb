```python
import pandas as pd

df_clock = pd.read_csv('C:/Users/CHARLIE/Desktop/文字探勘/期末/陳時中_社群網路節點.csv')
df_jiang = pd.read_csv('C:/Users/CHARLIE/Desktop/文字探勘/期末/蔣萬安_社群網路節點.csv')
df_33 = pd.read_csv('C:/Users/CHARLIE/Desktop/文字探勘/期末/黃珊珊_社群網路節點.csv')
```

```python
df = pd.concat([df_clock, df_jiang, df_33], axis = 0)
df
```

```python
df_node = pd.DataFrame({'Node':df['Target']})
df_node
```

```python
df_node.drop_duplicates(subset='Node', keep='first', inplace=True)
```

```python
df_node
```

```python
id_node = []
for i in range(len(df_node['Node'])):
    id_node.append(i+1)
df_node['Id'] = id_node
```

```python
df_node
```

```python
df_node.to_csv('C:/Users/CHARLIE/Desktop/文字探勘/期末/node.csv')
```

```python
node_id = dict(zip(df_node['Node'], df_node['Id']))
```

```python
node_id
```

```python
target = []
for i in df['Target']:
    target.append(node_id[i])
df['Target_id'] = target
```
